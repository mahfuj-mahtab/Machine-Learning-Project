```python
 def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df1[~df1['total_sqft'].apply(is_float)].head(50)
```

```python
def convert_sqrt_to_sum(x):
        token = x.split('-')
        if(len(token) == 2):
            return (float(token[0]) + float(token[1])) / 2
        try:
            return float(x)
        except:
            return None
```

```python
df2 = df1.copy()
```

```python
df2
```

```python
df2['total_sqft'] = df2['total_sqft'].apply(convert_sqrt_to_sum)
```
